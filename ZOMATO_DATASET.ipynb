#### 1) Analysing "rate" columns since ther are lot of null values present

```python
zomato_df['RATING'].unique()
```

```python
# Defining a function to handel the rating column and also covert them from text to integer form
def treat_ratings(values):
    if (values == 'NEW' or values == '-'):
        return np.nan                   ### NumPy NAN stands for not a number and 
                                   #    np.nan is defined as a substitute for declaring value which are numerical values that are missing values in an array as NumPy is used to deal with arrays in Python
    
    else :
        values = str(values).split('/')                # since split method splits string into list of int 
                                                       # i.e."4.1/5" will split it into 4.1 and /5 ---> list is =   values=[4.1][/5]
        values = values[0]                                # But here we only need numerator of rating because all the rating are out of 5
                                                       # So we have stated its index value.
        return float(values)                           # Since we want these values in floating form.
        
        
# Now applying the function on "rate" column        

zomato_df.RATING = zomato_df['RATING'].apply(treat_ratings)
```

```python
## Using fill na method 

zomato_df.RATING.fillna(zomato_df.RATING.mean(), inplace= True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'zomato_df' is your DataFrame

# Filter data for online and offline delivery separately
online_delivery_data = zomato_df[zomato_df['ONLINE'] == 'Yes']
offline_delivery_data = zomato_df[zomato_df['ONLINE'] == 'No']

# Calculate the quartiles and IQR for online delivery ratings
q1_online = online_delivery_data['RATING'].quantile(0.25)
q3_online = online_delivery_data['RATING'].quantile(0.75)
iqr_online = q3_online - q1_online

# Define upper and lower bounds to exclude outliers
upper_bound_online = q3_online + 1.5 * iqr_online
lower_bound_online = q1_online - 1.5 * iqr_online

# Filter outliers for online delivery
online_delivery_data_no_outliers = online_delivery_data[
    (online_delivery_data['RATING'] >= lower_bound_online) &
    (online_delivery_data['RATING'] <= upper_bound_online)
]

# Calculate statistics for online delivery ratings excluding outliers
min_online = online_delivery_data_no_outliers['RATING'].min()
max_online = online_delivery_data_no_outliers['RATING'].max()
average_online = online_delivery_data_no_outliers['RATING'].mean()

# Repeat the same process for offline delivery
q1_offline = offline_delivery_data['RATING'].quantile(0.25)
q3_offline = offline_delivery_data['RATING'].quantile(0.75)
iqr_offline = q3_offline - q1_offline

upper_bound_offline = q3_offline + 1.5 * iqr_offline
lower_bound_offline = q1_offline - 1.5 * iqr_offline

offline_delivery_data_no_outliers = offline_delivery_data[
    (offline_delivery_data['RATING'] >= lower_bound_offline) &
    (offline_delivery_data['RATING'] <= upper_bound_offline)
]

min_offline = offline_delivery_data_no_outliers['RATING'].min()
max_offline = offline_delivery_data_no_outliers['RATING'].max()
average_offline = offline_delivery_data_no_outliers['RATING'].mean()

# Create a figure and axis
fig, ax = plt.subplots(figsize=(7, 6))

# Create the boxplot
sns.boxplot(x=zomato_df['ONLINE'], y=zomato_df['RATING'], ax=ax)
ax.set_title('Restaurant Ratings by Online Delivery Availability')
ax.set_xlabel('Online Delivery Availability')
ax.set_ylabel('Ratings')

# Display statistics
print("Online Delivery Statistics (excluding outliers):")
print(f"Minimum Rating: {min_online}")
print(f"Maximum Rating: {max_online}")
print(f"Average Rating: {average_online}")

print("\nOffline Delivery Statistics (excluding outliers):")
print(f"Minimum Rating: {min_offline}")
print(f"Maximum Rating: {max_offline}")
print(f"Average Rating: {average_offline}")

# Show the plot
plt.show()
```

These observation are due to Outliers present in features

1) whether restaurants having online or not the average rating to restaurants given by the customers is 3.68434

2) maximum rating for online delivery of restaurants by people is around 4.5

3) minimum rating for online delivery given by customer is around 3.0

4) And when people go directly to restaurants they give maximum of ratings around 4.0

5) And when people go directly to restaurants they give manimum of ratings around 3.2

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'zomato_df' is your DataFrame

# Create a violin plot
ax = sns.violinplot(x=zomato_df['ONLINE'], y=zomato_df['RATING'])
plt.title('Restaurant Ratings by Online Delivery Availability')
plt.xlabel('Online Delivery Availability')
plt.ylabel('Ratings')

# Calculate statistics for online and offline delivery ratings
online_ratings = zomato_df[zomato_df['ONLINE'] == 'Yes']['RATING']
offline_ratings = zomato_df[zomato_df['ONLINE'] == 'No']['RATING']

min_online = online_ratings.min()
max_online = online_ratings.max()
average_online = online_ratings.mean()

min_offline = offline_ratings.min()
max_offline = offline_ratings.max()
average_offline = offline_ratings.mean()

# Annotate the plot with statistics
plt.annotate(f"Online Min: {min_online}", xy=(0, min_online), xytext=(-0.2, min_online + 0.1), fontsize=10, color='blue')
plt.annotate(f"Online Max: {max_online}", xy=(0, max_online), xytext=(-0.2, max_online + 0.1), fontsize=10, color='blue')
plt.annotate(f"Online Avg: {average_online:.4f}", xy=(0, average_online), xytext=(-0.2, average_online + 0.1), fontsize=10, color='blue')

plt.annotate(f"Offline Min: {min_offline}", xy=(1, min_offline), xytext=(1.05, min_offline + 0.1), fontsize=10, color='red')
plt.annotate(f"Offline Max: {max_offline}", xy=(1, max_offline), xytext=(1.05, max_offline + 0.1), fontsize=10, color='red')
plt.annotate(f"Offline Avg: {average_offline:.4f}", xy=(1, average_offline), xytext=(1.05, average_offline + 0.1), fontsize=10, color='red')

# Show the plot
plt.show()
```
